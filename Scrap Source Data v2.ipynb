```python
def readHtml(url):
    try:
        html = urlopen(url)
    except HTTPError as e:
        print(e)
        return None
    except URLError as e:
        print("The server could not be found!")
        return None
    else:
        return html
```

```python
def download_file(collection, page, date):
    if collection == "DCD":
        collection = "D"
    elif collection == "DCNR":
        collection = "R"
    elif collection == "DCN":
        collection = "J"
    elif collection == "DANC":
        collection = "R"
    elif collection == "ANA":
        collection = "A"
    if collection == "J":
        return
    file_url = "http://imagem.camara.gov.br/dc_20.asp?selCodColecaoCsv=" + collection
    file_url += "&txPagina=" + page
    file_url += "&Datain=" + date + "&txSuplemento="
    print(file_url)
    response = requests.get(file_url)
    r = re.search(r'http://Imagem.camara.gov.br/Imagem/[drja]+/pdf/[\w]+.pdf', response.text)
    if r is not None:
        pdf_url = r.group()
        print(pdf_url)
        download_content = requests.get(pdf_url)
        if download_content is not None:
            with open(collection + "_" + page + "_" + date.replace("/", "-") + ".pdf",'wb') as f:
                f.write(download_content.content)
```

```python
def get_speech_from_html(source, html):
    title = ""
    if html.title is not None:
        title = html.title.get_text().encode(encoding='UTF-8').decode(encoding='UTF-8')
    summary = ""
    if html.txSumarioID is not None:
        summary = html.txSumarioID.get_text().encode(encoding='UTF-8').decode(encoding='UTF-8')
    rows = get_table_rows(html)
    cols = row.find_all("td")
    date = cols[0].getText().encode(encoding='UTF-8').decode(encoding='UTF-8')
    session = cols[1].getText().encode(encoding='UTF-8').decode(encoding='UTF-8')
    phase = cols[2].getText().encode(encoding='UTF-8').decode(encoding='UTF-8')
    speech = ""
    if html.p is not None:
        speech = html.p.getText().encode(encoding='UTF-8').decode(encoding='UTF-8')
    result = {
        'source': source,
        'title': title,
        'date': date,
        'session': session,
        'phase': phase,
        'summary': summary,
        'speech': speech
    }  
    return result
```
