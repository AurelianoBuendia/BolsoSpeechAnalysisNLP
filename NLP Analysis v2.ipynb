```python
file = open('./data/speeches.json', encoding='utf-8')
if file is not None:
    items = json.load(file)
    file.close()
else:
    print("Error reading data file.")
```

```python
speeches = []
for item in items:
    speeches.append(item['speech'])
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
import re
import nlpnet
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese') 
```

```python
def cleanup_sentence(sentence):
    result = re.sub(r'(?<=[.,])(?=[^\s])', r' ', sentence)
    return result
```

```python
def cleanup_word(word):
    word = word.strip('\'')
    result = re.sub(r'[^\w\s]', '', word)
    return result
```

```python
tagger = nlpnet.POSTagger('./nlpnet_models/pos-pt/', language='pt')
```

```python
bolso_words = set()
bolso_sentences = []
for speech in speeches:
    speech = cleanup_sentence(speech)
    sentences = sent_tokenize(speech)
    for sentence in sentences:
        bolso_sentences.append(sentence)
    # word_tokenize() returns a list of words, so the next line gives a list of lists of words, one per sentence
    word_lists = [word_tokenize(sentence) for sentence in sentences]
    for word_list in word_lists:
        for word in word_list:
                bolso_words.add(word)
bolso_words = list(bolso_words)
bolso_words = [cleanup_word(word) for word in bolso_words if cleanup_word(word) != ""]
```

```python
full_text = ""
for speech in speeches:
    full_text += "\n" + speech
#with open("./bolso_corpus.txt", 'w', encoding='utf8') as f:
#    f.write(full_text)
freq_dist = nltk.FreqDist([word.lower() for word in full_text])
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_wordcloud(data, title=None, stopwords = nltk.corpus.stopwords.words('portuguese')):
    wordcloud = WordCloud(
    background_color='white',
    stopwords=stopwords,
    max_words=100,
    max_font_size=40,
    scale=3,
    random_state=1).generate(' '.join(data))
    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title:
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)
    plt.imshow(wordcloud)
    plt.show()
```

```python
show_wordcloud(bolso_words)
```

```python
words_by_tags = {}
for tags in bolso_tags:
    for tag_tuples in tags:
        for tag_tuple in tag_tuples:
            key = tag_tuple[1]
            word = tag_tuple[0]
            if key in words_by_tags:
                words_by_tags[key].append(word)
            else:
                words_by_tags[key] = list()
```
